```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Plotting a Linear Function
x = np.linspace(-10, 10, 400)
y = 2 * x + 3
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x + 3')
plt.title('Linear Function Plot')
plt.legend()
plt.grid(True)
plt.show()

# 2. Plotting Multiple Functions
x = np.linspace(0, 10, 400)
y1 = x**2
y2 = np.sqrt(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = x^2', linestyle='--')
plt.plot(x, y2, label='y = sqrt(x)', color='red')
plt.title('Multiple Functions Plot')
plt.legend()
plt.grid(True)
plt.show()

# 3. Plotting Trigonometric Functions
x = np.linspace(0, 2*np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = sin(x)')
plt.plot(x, y2, label='y = cos(x)')
plt.plot(x, y3, label='y = tan(x)')
plt.title('Trigonometric Functions Plot')
plt.legend()
plt.grid(True)
plt.ylim(-2, 2)  # Limit y-axis to avoid large values due to tan(x)
plt.show()

# 4. Plotting Exponential Growth
x = np.linspace(-5, 5, 400)
y = np.exp(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = e^x')
plt.title('Exponential Growth Plot')
plt.legend()
plt.grid(True)
plt.show()

# 5. Plotting Piecewise Functions
x = np.linspace(-2, 2, 400)
y = np.piecewise(x, [x < 0, x >= 0], [-1, 1])
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Piecewise Function')
plt.title('Piecewise Function Plot')
plt.legend()
plt.grid(True)
plt.show()

# 6. Plotting Parametric Equations
t = np.linspace(0, 2*np.pi, 400)
x = np.cos(t)
y = np.sin(t)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Parametric Equations')
plt.title('Parametric Equations Plot')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensure equal scaling for both axes
plt.show()


# 7. Plotting Logarithmic Functions
x = np.linspace(0.1, 10, 400)
y1 = np.log(x)
y2 = np.log10(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='y = ln(x)')
plt.plot(x, y2, label='y = log10(x)', linestyle='--')
plt.title('Logarithmic Functions Plot')
plt.legend()
plt.grid(True)
plt.show()

# 8. Plotting Step Function
x = np.linspace(-5, 5, 400)
y = np.piecewise(x, [x < 0, x >= 0], [0, 1])
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Step Function')
plt.title('Step Function Plot')
plt.legend()
plt.grid(True)
plt.show()

# 9. Plotting Polynomial Functions
x = np.linspace(-3, 3, 400)
y = x**3 - 2*x**2 + x - 1
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Polynomial Function')
plt.title('Polynomial Function Plot')
plt.legend()
plt.grid(True)
plt.show()

# 10. Plotting Piecewise Linear Functions
x = np.linspace(-2, 2, 400)
y = np.piecewise(x, [x <= 0, x > 0], [lambda x: -x, lambda x: x])
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Piecewise Linear Function')
plt.title('Piecewise Linear Function Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    def __init__(self, length, width):

        pass

    @abstractmethod

    def area(self):

        return self.length * self.width

 

class Rectangle(Shape):

    def area(self):

        return self.length * self.width

 

r = Rectangle(5, 4)

print(r.area())
```
