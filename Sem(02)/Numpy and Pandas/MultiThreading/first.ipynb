````python

def fun1():
    for i in range (1,11):
        print(i)
        time.sleep(3)

def fun2():
    for i in range(65,91):
        print(chr(i))
        time.sleep(.1)
        
````

````python
t1 = threading.Thread(target = fun1)
t2 = threading.Thread(target = fun2)
t1.start()
t2.start()
````
